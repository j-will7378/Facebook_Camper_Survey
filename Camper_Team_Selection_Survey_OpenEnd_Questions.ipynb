```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Load survey data
# Replace 'your_file.csv' and 'open_end_column' with the actual file path and column name
# data = pd.read_csv('your_file.csv')
responses = data['Why are you not satisfied with the team selection process?']

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Text preprocessing
stop_words = set(stopwords.words('english'))
preprocessed_responses = responses.apply(lambda x: ' '.join([word.lower() for word in nltk.word_tokenize(str(x)) if word.isalpha() and word.lower() not in stop_words]))
```

```python
# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the preprocessed responses
tfidf_matrix = vectorizer.fit_transform(preprocessed_responses)
```

```python
# Define the number of clusters
num_clusters = 3

# Initialize the K-Means model
km = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the model
km.fit(tfidf_matrix)

# Get the cluster assignments for each response
clusters = km.labels_.tolist()
```

```python
# Add cluster assignments to the original DataFrame
data['cluster'] = clusters
# Set the display option to show the entire content of a column
pd.set_option('display.max_colwidth', None)

# Examine the content of each cluster
for i in range(num_clusters):
    print(f"Cluster {i}:")
    print(data[data['cluster'] == i]['Why are you not satisfied with the team selection process?'])
    print()
```

# Observation
**Whay campeers are not satisfied**
- Geo restricting can be hard.
- Confusion on the availability of teams.
- The tool feels useless if teams are not reaching out.
- The manger experience is different, but they go through the process a like an IC.

# Topic Modeling

```python
# preprocess data
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Load survey data
# Replace 'your_file.csv' and 'text_column_name' with the actual file path and column name
data = pd.read_csv('/content/Camper-Team-Selection-Survey_OpenEnd_Survey_Questions_2 - Sheet1 (1).csv')
responses = data['Why are you not satisfied with the team selection process?']

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Define stopwords
stop_words = set(stopwords.words('english'))

# Preprocess the text data
preprocessed_responses = responses.apply(lambda x: ' '.join([word.lower() for word in nltk.word_tokenize(str(x)) if word.isalpha() and word.lower() not in stop_words]))
```

```python
# Vectorize Text Data Convert the preprocessed text data into a document-term matrix using CountVectorizer.
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Fit and transform the preprocessed responses
term_matrix = vectorizer.fit_transform(preprocessed_responses)
```

```python
# Apply LDA Model Fit the LDA model to the document-term matrix to identify topics.
from sklearn.decomposition import LatentDirichletAllocation

# Define the number of topics you want to identify
n_topics = 5

# Initialize and fit the LDA model
lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda_model.fit(term_matrix)
```

```python
# Display Topics Once the model is trained, display the top words for each identified topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Define the number of top words you want to display for each topic
no_top_words = 10

# Get feature names (words) from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# Display the topics and their top words
display_topics(lda_model, feature_names, no_top_words)
```

```python
# Assign the most relevant topic to each response
data['topic'] = lda_model.transform(term_matrix).argmax(axis=1)

# View the distribution of topics
print(data['topic'].value_counts())
```
